# 5. mājās darbs

### 1. Izveidot izteiksmi un aprēķināt to

```python
from sympy import cos, cot, exp, integrate, limit, oo, pi, simplify, sin, solve, symbols
from sympy import Eq, Matrix, Rational
from sympy.plotting import plot
```

Izveidot izteiksmi $f = xe^{-x} + x(1-x)$

```python
x = symbols('x')
f = x*exp(-x) + x*(1 - x)
f
```

Attēlot izteiksmi

```python
plot(f, (x, -2, 7))
```

Aprēķināt izteiksmi sekojošām $x$ vērtībām $x = 0, 1, 2.5, 3.1, 4.5, 5$ un izvadīt visus rezultātus

```python
x_values = [0, 1, 2.5, 3.1, 4.5, 5]
results = [f.subs(x, val) for val in x_values]

for x, y in zip(x_values, results):
    print(f"f({x}) = {y}")
```

### 2. Mērvienības

Otrais Ņūtona likums nosaka to, ka ķermeņa paātrinājums ir tieši proporcionāls spēkam, kas uz to darbojas, un apgriezti proporcionāls masai, kas piemīt ķermenim. Tā kā uz ķermeni reizē var darboties vairāki spēki, tad Ņūtona likumā esošo spēku sauc par kopspēku. Ņūtona otro likumu var izteikt ar sekojošu formulu: $F = ma$. Aprēķināt spēku pie masas $12k$ un paātrinājuma $10m/s^2$ un konvertēt rezultātu ņūtonos.

```python
m, a, F = symbols('m a F')
force_equation = Eq(F, m*a)

m_value = 12
a_value = 10

result = force_equation.subs({m: m_value, a: a_value})
result
```

Izvadīt rezultātu ar pavadošo paskaidrojuma tekstu

```python
print(f"Spēks pie masas {m_value}kg un paātrinājuma {a_value}m/s^2 ir {result.rhs} ņūtoni.")
```

### 3. Faktorizēšana un sakņu atrašana

```python
x = symbols('x')
pol = x**4 - 6*x**3 + x**2 + 24*x + 16
pol
```

Faktorizēt to un attēlot rezultātu.

```python
factorized_pol = pol.factor()
factorized_pol
```

Atrast saknes un pārbaudīt katras saknes skaitu

```python
roots = solve(factorized_pol, x)
roots_count = {root: factorized_pol.as_poly().count_roots(root) for root in roots}

for root, count in roots_count.items():
    print(f"Root {root} gets repeated {count} times.")
```

**Remark:** It seems that root $x=4$ should be repeated twice, but SymPy identifies it only once...

```python
plot(factorized_pol, (x, -2, 5))
```

### 4. Atvasināšana un itegrēšana

Izveidot, saglabāt $(f)$ un attēlot funkciju $x^4 - 6x^3 + x^2 + 24x + 16$

```python
x = symbols('x')
f = x**4 - 6*x**3 + x**2 + 24*x + 16
plot(f, (x, -5, 8))
```

Aprēķināt funkcijas vērtību pie $x = 2$

```python
f_value_at_2 = f.subs(x, 2)
f_value_at_2
```

Atrast, saglabāt (f_int) un attēlot funkcijas integrāli

```python
f_int = integrate(f, x)
plot(f_int, (x, -5, 8))
```

Aprēķināt integrāļa vērtību pie $x = 2$

```python
f_int_value_at_2 = f_int.subs(x, 2)
f_int_value_at_2
```

Atrast, saglabāt (f_int_atv) un attēlot iepriekš iegūtā integrāļa atvasinājumu

```python
f_int_dif = f_int.diff(x)
plot(f_int_dif, (x, -5, 8))
```

Aprēķināt atvasinājuma vērtību pie $x = 2$

```python
f_int_dif_value_at_2 = f_int_dif.subs(x, 2)
f_int_dif_value_at_2
```

Attēlot un salīdzināt funkciju (f) ar tās integrāļa atvasinājumu (f_int_atv)

```python
plot(f, f_int_dif, (x, -5, 8), legend=True)
```

Izveidot izteiksmi $e^{\cos(x)} + \frac{1}{3}$

# 6. Atvasināšana un tās praktiskais pielietojums

Izveidot funkciju $x^3 + x^2 + 1$

```python
x = symbols('x')
f = x**3 + x**2 + 1
f
```

Uzzimēt funkcijas grafiku
f

```python
p = plot(f, (x, -1.6, 1.2))
```

Analītiski atrast funkcijas lokālo minimumu un maksimumu (izmantot atvasinājumu)

```python
f_1d = f.diff(x)
critical_points = solve(Eq(f_1d, 0))
critical_points
```

Izmantojot atvasinājumus identificēt ekstrēmu veidu - lokālais minimums, lokālais maksimums vai pārliekuma punkts

```python
f_2d = f_1d.diff(x)

for point in critical_points:
    if f_2d.subs(x, point) > 0:
        print(f"{point} is a local minimum")
    elif f_2d.subs(x, point) < 0:
        print(f"{point} is a local maximum")
    else:
        print(f"{point} is a saddle point")
```

Izdrukāt funkcijas lokālā minimuma un maksimuma vērtības

```python
for point in critical_points:
    p.extend(plot(f.subs(x, point), (x, -1.6, 1.2), show=False))
    print(f"f({point}) = {f.subs(x, point)}")
```

Iezīmēt funkcijas grafikā ar horizontālu taisni funkcijas lokālo minimumu un maksimumu

```python
p.show()
```

# 9. Lineāru vienādojumu sistēmas risināšana

Atrisini lineāro vienādojumu sistēmu:
\[
\begin{align*}
2x + 3y &= 7 \\
4x - y &= 6
\end{align*}
\]

```python
x, y = symbols('x y')

eq1 = Eq(2*x + 3*y, 7)
eq2 = Eq(4*x - y, 6)

solution = solve((eq1, eq2), (x, y))
solution
```

# 10. Lineārā vienādojuma sistēmas risināšana izmantojot inverso matricu

Izveidot vienādojumu:
\[
\begin{align*}
2x + 3y &= z \\
-1x - 4y &= 2z
\end{align*}
\]

```python
x, y, z = symbols('x y z')

eq1 = Eq(2*x + 3*y, z)
eq2 = Eq((-1)*x - 4*y, 2*z)
```

Aprēķināt lineārās sistēmas vienādojumu izmantojot inverso matricu $x = A^{-1} \times b$

```python
A = Matrix([[2, 3], [-1, -4]])
b = Matrix([[z], [2*z]])

A_inv = A.inv()

x = A_inv * b
x
```

Aprēķināt un izdrukāt rezultātu pie $z$ vērtības vienādas ar $2$

```python
x.subs(z, 2)
```
